```python
import pandas as pd
import matplotlib.pyplot as plt

# load CSV file
data = pd.read_csv('./data/points7.csv')

x = data['x']
y = data['y']

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of data')
plt.show()
```

```python
import numpy as np

a_values = np.linspace(0,3,100)

losses = []
for a in a_values:
    loss = np.mean((a * x - y) ** 2)
    losses.append(loss)

plt.plot(a_values, losses)
plt.xlabel('a')
plt.ylabel('Loss')
plt.title('Loss function')
plt.grid(True) 
plt.show()  
```

```python
def loss_fn(x : np.ndarray, y : np.ndarray, a : float) -> float:
    return np.mean((a * x - y) ** 2)


a = 3.0
h = 0.001

print(loss_fn(x, y, a))
print(loss_fn(x, y, a - h))
print(loss_fn(x, y, a + h))
```
